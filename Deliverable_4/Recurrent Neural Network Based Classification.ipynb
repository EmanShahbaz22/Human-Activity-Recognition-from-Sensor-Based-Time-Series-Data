```python
!pip install -q tensorflow scikit-learn

from google.colab import files
uploaded = files.upload() 
zip_filename = list(uploaded.keys())[0]
print("Uploaded:", zip_filename)

import zipfile, os
with zipfile.ZipFile(zip_filename, 'r') as z:
    z.extractall("/content/")
print("Extracted. Exists:", os.path.exists("/content/UCI HAR Dataset"))

import numpy as np
import pandas as pd

def load_signal(path):
    return np.loadtxt(path)

X_train = np.dstack([
    load_signal("/content/UCI HAR Dataset/train/Inertial Signals/body_acc_x_train.txt"),
    load_signal("/content/UCI HAR Dataset/train/Inertial Signals/body_acc_y_train.txt"),
    load_signal("/content/UCI HAR Dataset/train/Inertial Signals/body_acc_z_train.txt"),
    load_signal("/content/UCI HAR Dataset/train/Inertial Signals/body_gyro_x_train.txt"),
    load_signal("/content/UCI HAR Dataset/train/Inertial Signals/body_gyro_y_train.txt"),
    load_signal("/content/UCI HAR Dataset/train/Inertial Signals/body_gyro_z_train.txt")
])

X_test = np.dstack([
    load_signal("/content/UCI HAR Dataset/test/Inertial Signals/body_acc_x_test.txt"),
    load_signal("/content/UCI HAR Dataset/test/Inertial Signals/body_acc_y_test.txt"),
    load_signal("/content/UCI HAR Dataset/test/Inertial Signals/body_acc_z_test.txt"),
    load_signal("/content/UCI HAR Dataset/test/Inertial Signals/body_gyro_x_test.txt"),
    load_signal("/content/UCI HAR Dataset/test/Inertial Signals/body_gyro_y_test.txt"),
    load_signal("/content/UCI HAR Dataset/test/Inertial Signals/body_gyro_z_test.txt")
])

y_train = pd.read_csv("/content/UCI HAR Dataset/train/y_train.txt", header=None)[0].values - 1
y_test  = pd.read_csv("/content/UCI HAR Dataset/test/y_test.txt", header=None)[0].values - 1

print("X_train:", X_train.shape, " y_train:", y_train.shape)
print("X_test :", X_test.shape,  " y_test :", y_test.shape)

mu = X_train.mean(axis=(0,1), keepdims=True)
sd = X_train.std(axis=(0,1), keepdims=True) + 1e-8

X_train_n = (X_train - mu) / sd
X_test_n  = (X_test  - mu) / sd

print("Normalized shapes:", X_train_n.shape, X_test_n.shape)

# SimpleRNN, LSTM, GRU
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, GRU, Dense, Dropout
import matplotlib.pyplot as plt

tf.random.set_seed(42)
np.random.seed(42)

TIMESTEPS = X_train_n.shape[1]  
CHANNELS  = X_train_n.shape[2] 
N_CLASSES = 6

def build_model(model_type="rnn"):
    model = Sequential()
    model.add(tf.keras.Input(shape=(TIMESTEPS, CHANNELS)))

    if model_type == "rnn":
        model.add(SimpleRNN(64))
    elif model_type == "lstm":
        model.add(LSTM(64))
    elif model_type == "gru":
        model.add(GRU(64))
    else:
        raise ValueError("model_type must be: rnn, lstm, gru")

    model.add(Dropout(0.3))
    model.add(Dense(64, activation="relu"))
    model.add(Dense(N_CLASSES, activation="softmax"))

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model

from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report

def plot_loss(history, title):
    plt.figure()
    plt.plot(history.history["loss"], label="Train Loss")
    plt.plot(history.history["val_loss"], label="Val Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(title)
    plt.legend()
    plt.show()

def plot_cm(cm, title):
    plt.figure()
    plt.imshow(cm)
    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.colorbar()
    plt.show()

def train_and_evaluate(model_type, epochs=20, batch_size=64):
    name_map = {"rnn":"Simple RNN", "lstm":"LSTM", "gru":"GRU"}
    name = name_map[model_type]

    model = build_model(model_type)
    print("\n==============================")
    print(f"Training: {name}")
    print("==============================")
    model.summary()

    history = model.fit(
        X_train_n, y_train,
        validation_split=0.2,
        epochs=epochs,
        batch_size=batch_size,
        verbose=1
    )

    plot_loss(history, f"{name}: Training vs Validation Loss")

    y_proba = model.predict(X_test_n, verbose=0)
    y_pred = np.argmax(y_proba, axis=1)

    acc = accuracy_score(y_test, y_pred)
    f1w = f1_score(y_test, y_pred, average="weighted")
    cm = confusion_matrix(y_test, y_pred)

    print(f"\n{name} TEST METRICS")
    print(f"Accuracy: {acc:.4f}")
    print(f"Weighted F1: {f1w:.4f}")
    print("\nConfusion Matrix:")
    print(pd.DataFrame(cm))
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, zero_division=0))

    plot_cm(cm, f"{name}: Confusion Matrix")

    return {
        "Model": name,
        "Accuracy": acc,
        "F1_weighted": f1w,
        "ConfusionMatrix": cm
    }

results = []
results.append(train_and_evaluate("rnn",  epochs=20))
results.append(train_and_evaluate("lstm", epochs=20))
results.append(train_and_evaluate("gru",  epochs=20))

# Comparison table of all recurrent models
results_df = pd.DataFrame([{
    "Model": r["Model"],
    "Accuracy": r["Accuracy"],
    "Weighted F1": r["F1_weighted"]
} for r in results])

print("\n\n=== RNN MODEL COMPARISON (5.4) ===")
print(results_df)

#  Compare with FCNN (5.3)
fcnn_acc_53 = 0.9437
print("\n=== COMPARISON WITH FCNN (5.3) ===")
print(f"FCNN Accuracy (5.3): {fcnn_acc_53:.4f}")
print("RNN Models (5.4):")
print(results_df)

best_rnn_acc = results_df["Accuracy"].max()
best_rnn_name = results_df.loc[results_df["Accuracy"].idxmax(), "Model"]
print(f"\nBest recurrent model: {best_rnn_name} with Accuracy={best_rnn_acc:.4f}")
print(f"Difference (Best RNN - FCNN): {(best_rnn_acc - fcnn_acc_53):.4f}")
```
